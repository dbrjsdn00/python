```python
t=np.arange(0,12,0.5)
%matplotlib inline
y1=np.sin(t)
y2=np.cos(t)
plt.plot(t, y1, 'y--')
plt.plot(t, y2, 'b--', marker='*')
plt.xlabel('time')
plt.ylabel('amp')
plt.title('SinCos')
plt.legend(['Sin','Cos'])
plt.grid()
```

## [문제] subplot 을 이용해서 2행 1열 구조로 2개씩 / 4가지 색상 다르게 / 4가지 스타일 다르게 / 마커, 범례표시 / x,y 제목 / x=0부터 5까지 단위1 / y1=x+1, y2=x+2, y3=x+3, y4=x+4

```python
x=np.arange(0,5,1)
y1=x+1
y2=x+2
y3=x+3
y4=x+4
plt.subplot(1,2,1)
plt.plot(x,y1, 'r--', marker='v')
plt.plot(x,y2, 'g--', marker='p')
plt.subplot(1,2,2)
plt.plot(x,y3, 'b--', marker='*')
plt.plot(x,y4, 'y--', marker='^')
plt.xlabel('time')
plt.ylabel('amp')
```
